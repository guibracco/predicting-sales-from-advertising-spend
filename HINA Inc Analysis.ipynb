**Assignment:** Predicting Sales from Advertising Spend.

**Problem Statement:**
HINA Inc. is a large FMCG company that is streamlining its marketing budget. This involves taking stock of all its marketing strategies. This, in turn, means re-assessing the effectiveness of its existing spend on various marketing channels. As a marketing data analyst, you need to figure out if spending money on TV advertising campaigns results in a direct increase in sales. In other words, you need to find out if the TV advertising spend and the sales figures share a linear relationship and how strong that relationship is.

**Data:** You are provided with historical advertising data – weekly sales and spend on each channel – for almost the 4 previous years.

Assuming an alpha value of 0.05 or 5%, your analysis should state clearly if there is significant evidence to show that the TV advertising campaigns have a relationship with the sales.

**FURTHER WORK:** Perform the same analysis/operations for the Radio and Newspaper campaigns and state resulting feedback.

**SUBMISSION:** A PYTHON SCRIPT detailing data ingestion, cleaning (if need - Explain why any method of cleaning or replacement was used), analysis between the Individual campaign spends and the Sales.

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

First, let's load and check the data for missing values. Any non-numeric values will be converted to NaN and considered missing.

```python
df = pd.read_csv("advertising.csv")

# Convert any non-numeric value to NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Get a quick view of the table
display(df)

# Is any value missing?
check_nan = df.isnull().values.any()
if not check_nan:
    print("No value is missing.")
else:
    print("Some value is missing!")
```

TV: Amount spent on TV ads.  
Radio: Amount spent on radio ads.  
Newspaper: Amount spent on newspaper ads.  
Sales: Return in sales.  

The amounts are in thousands of dollars and the data were gathered weekly.

Lets calculate the Pearson coeficient correlation between each of ads category and the sales.

```python
tv_and_sales = df.Sales.corr(df.TV)
radio_and_sales = df.Sales.corr(df.Radio)
news_and_sales = df.Sales.corr(df.Newspaper)

print("The correlation between TV ads and sales is: ", tv_and_sales)
print("The correlation between radio ads and sales is: ", radio_and_sales)
print("The correlation between newspaper ads and sales is: ", news_and_sales)
```

There's a clear and strong correlation between TV ads and sales.  
The correlation for radio ads is moderate at best.  
The correlation for newspaper ads is very weak.

Time for some visualization. Let's plot the data using sns.regplot() for a linear regression with CI of 95% (alpha of 5%).

Let's also test the hypothesis of non-correlation with a p-value.

```python
# Sales versus TV ads
sns.regplot(x='TV', y='Sales', data=df, ci=95)
_ = plt.title("Correlation of TV Ads and Sales")
_ = plt.xlabel("TV Ads (thousands of dollars)")
_ = plt.ylabel("Sales (thousands of dollars)")

p_coef, p_value = stats.pearsonr(df['TV'], df['Sales'])
print("Pearson correlation coefficient:\t", p_coef)
print("P-value:\t\t\t\t", p_value)
```

It's clear now that there's a strong and positive correlation between the amount spent in TV ads and return in sales. Not just that, but the CI of 95% also shows us that even our bootstrapping samples are narrowly concentrated around the estimated linear correlation. Futhermore, p-value is extremely small for us to discard the correlation.

```python
# Sales versus Radio ads
sns.regplot(x='Radio', y='Sales', data=df, ci=95)
_ = plt.title("Correlation of Radio Ads and Sales")
_ = plt.xlabel("Radio Ads (thousands of dollars)")
_ = plt.ylabel("Sales (thousands of dollars)")

p_coef, p_value = stats.pearsonr(df['Radio'], df['Sales'])
print("Pearson correlation coefficient:\t", p_coef)
print("P-value:\t\t\t\t", p_value)
```

The correlation between Radio ads and sales is a moderate one and therefore less consistent than what's observed with TV ads. Again, p-value is too small to discard the correlation.

```python
# Sales versus Newspaper ads
sns.regplot(x='TV', y='Newspaper', data=df, ci=95)
_ = plt.title("Correlation of Newspaper Ads and Sales")
_ = plt.xlabel("News Ads (thousands of dollars)")
_ = plt.ylabel("Sales (thousands of dollars)")

p_coef, p_value = stats.pearsonr(df['Newspaper'], df['Sales'])
print("Pearson correlation coefficient:\t", p_coef)
print("P-value:\t\t\t\t", p_value)
```

The correlation here is too week and the effect of Newspaper ads doesn't seem to be relevant for Sales. However, p-value is yet sufficiently small (less than 5%) for us to discard the correlation.

Conclusion:

Although sales are deffinitely and positively correlated to ads, especially TV ads, more data would be necessary to analyse the effect that each media has on revenue separately and over time. Although correlation is not necessarily causation, we lack the data to determine if there's another factor affecting sales.
